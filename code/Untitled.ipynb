```python
url = 'https://gutenberg.org/cache/epub/1597/pg1597.txt' 
```

```python
url
```

```python
from urllib.request import urlopen

with urlopen(url) as response:
    text_string = response.read().decode('utf-8')
```

```python
start = text_string.find('CONTENTS') + len('CONTENTS')

end = text_string.find('*** END OF THE PROJECT')
```

```python
start
```

```python
end
```

```python
content = text_string[start:end]
```

```python
length = len(content)
```

```python
lower_text = content.lower()
```

```python
clean_text = lower_text.replace('\r', ' ').replace('\n', ' ').replace(',', ' ').replace(';', ' ')
```

```python
clean_text
```

```python
word = 'bird'
word_count = clean_text.count(word)
word_count
```

```python
word_list = clean_text.split()
```

```python
words_in_text = len(word_list)
```

```python
words_in_text
```

```python
res = round(word_count / words_in_text * 100, 2)
```

```python
print (f'The word {word} is {word_count} times in the text')
```
